```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import load_breast_cancer
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score

# Veri kümesini yükle
data = load_breast_cancer()
X = data.data
y = data.target

# Veriyi eğitim ve test setlerine bölelim
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Lojistik Regresyon modelini eğitelim
model = LogisticRegression(max_iter=5000)
model.fit(X_train, y_train)

# Test verisi üzerinde tahminler yapalım
y_pred = model.predict(X_test)

# Doğruluk değerini hesaplayalım
accuracy = accuracy_score(y_test, y_pred)
print(f"Model accuracy: {accuracy:.2f}")

# Görselleştirme için ilk iki özelliği kullanalım
plt.scatter(X_test[y_test == 0][:, 0], X_test[y_test == 0][:, 1], color="red", label="Malignant")
plt.scatter(X_test[y_test == 1][:, 0], X_test[y_test == 1][:, 1], color="green", label="Benign")

# Tahminlerin renklerini belirleyelim
colors = np.where(y_pred == 1, "lime", "darkred")

# Tahmin noktalarını çizdirelim
plt.scatter(X_test[:, 0], X_test[:, 1], c=colors, marker="x", label="Predictions")

# Eksen etiketleri ve başlık ekleyelim
plt.xlabel(data.feature_names[0])
plt.ylabel(data.feature_names[1])
plt.title("Breast Cancer Classification")

# Efsane ekleyelim
plt.legend()

# Grafiği gösterelim
plt.show()
```
